```python

# QFT, INV QFT, DIFUSOR GROVER

def qft(qc, qubits):
    n = len(qubits)
    for i in range(n):
        for j in range(i):
            qc.cp(np.pi / 2**(i-j), qubits[j], qubits[i])
        qc.h(qubits[i])
    for i in range(n // 2):
        qc.swap(qubits[i], qubits[n - i - 1])

def inverse_qft(qc, qubits):
    n = len(qubits)
    for i in range(n // 2):
        qc.swap(qubits[i], qubits[n - i - 1])
    for i in reversed(range(n)):
        for j in reversed(range(i)):
            qc.cp(-np.pi / 2**(i-j), qubits[j], qubits[i])
        qc.h(qubits[i])

def grover_diffuser(qc, qubits):
    qubit_list = list(qubits)

    # H en todos
    for q in qubit_list:
        qc.h(q)
    # X en todos
    for q in qubit_list:
        qc.x(q)

    # Multi-controlled Z
    qc.h(qubit_list[-1])
    qc.mcx(qubit_list[:-1], qubit_list[-1])
    qc.h(qubit_list[-1])

    # X en todos
    for q in qubit_list:
        qc.x(q)
    # H en todos
    for q in qubit_list:
        qc.h(q)
```

```python

# CONSTRUIR EL CIRCUITO

def build_quantum_self_attention_circuit(inputs):
    qubits = 8
    qc = QuantumCircuit(qubits)

    # Basis Embedding
    binarized = [1 if x >= 0 else 0 for x in inputs]
    for i, b in enumerate(binarized):
        if b == 1:
            qc.x(i)

    # 3 capas
    n_layers = 3
    for layer in range(n_layers):
        for i in range(qubits):
            angle_rx = inputs[i % 8] * (layer + 1)
            angle_ry = inputs[(i + 1) % 8] * (layer + 1)
            angle_rz = inputs[(i + 2) % 8] * (layer + 1)
            qc.rx(angle_rx, i)
            qc.ry(angle_ry, i)
            qc.rz(angle_rz, i)

        for i in range(qubits - 1):
            angle_crz = np.pi / (layer + 2)
            qc.crz(angle_crz, i, i + 1)
            qc.cx(i, i + 1)   # cx en lugar de cnot

    # QFT + inverse QFT + Grover
    qft(qc, range(qubits))
    inverse_qft(qc, range(qubits))
    grover_diffuser(qc, range(qubits))

    # Operaciones finales
    for i in range(qubits):
        qc.h(i)
        qc.t(i)
        qc.rz(inputs[i % 8], i)

    # Re-embeding
    for i, b in enumerate(binarized):
        if b == 1:
            qc.x(i)

    # guardar statevector
    qc.save_statevector()

    return qc

def quantum_self_attention(inputs):
    qubits = 8
    qc = build_quantum_self_attention_circuit(inputs)

    # Simulador en modo statevector
    simulator = AerSimulator(method='statevector')

    # Transpilar
    qc_compiled = transpile(qc, simulator)

    # Ejecutar
    job = simulator.run(qc_compiled, shots=1)
    res = job.result()

    # Recuperar el estado final
    final_state = res.data(0)['statevector']

    # Calcular <Z> en cada qubit
    z_values = []
    for i in range(qubits):
        label = ['I'] * qubits
        label[qubits - 1 - i] = 'Z'
        op_z = Operator.from_label("".join(label))

        # Convertir final_state en objeto Statevector
        sv = Statevector(final_state)
        expval = sv.expectation_value(op_z)
        z_values.append(np.real(expval))

    return z_values
```

```python
# SOFTMAX

def softmax(x):
    x = np.array(x, dtype=float)
    exps = np.exp(x)
    return exps / np.sum(exps)
```

```python
# GENERAR FRASE
def GenerateSentence(input_):
    tokens = word_tokenize(input_.lower())

    # Embeddings de palabras conocidas
    embed_list = []
    for w in tokens:
        if w == "_":  # _ para indicar huecos a rellenar
            embed_list.append(None)
        elif w in word2vec.wv:
            embed_list.append(word2vec.wv[w])
        else:
            embed_list.append(np.zeros(8, dtype=float))

    gen_words = []
    for emb in embed_list:
        if emb is None:  # Palabra a generar
            known_embs = [e for e in embed_list if e is not None]
            if known_embs:
                avg_emb = np.mean(known_embs, axis=0)
            else:
                avg_emb = np.zeros(8, dtype=float)

            att_out = quantum_self_attention(avg_emb)
            att_soft = softmax(att_out)
            idx = int(np.argmax(att_soft))
            if idx < len(word2vec.wv.index_to_key):
                chosen_w = word2vec.wv.index_to_key[idx]
            else:
                chosen_w = "desconocida"
            gen_words.append(chosen_w)
        else:
            # Palabra conocida
            idx = word2vec.wv.similar_by_vector(emb, topn=1)[0][0]
            gen_words.append(idx)

    return " ".join(gen_words)


```
